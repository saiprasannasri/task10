```python
import pandas as pd
import numpy as np
df=pd.read_csv("/content/student performance.csv")
```

```python
df.info()
```

```python
print("Shape:", df.shape)
df.head()

```

```python
df.describe()
```

```python
missing_percent = (df.isnull().sum() / len(df)) * 100
print(missing_percent)
```

```python
# Numerical columns → fill with median
num_cols = df.select_dtypes(include=np.number).columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())

# Categorical columns → fill with mode
cat_cols = df.select_dtypes(include='object').columns
for col in cat_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
import matplotlib.pyplot as plt
df[num_cols].hist(figsize=(12, 8))
plt.show()
```

```python
plt.figure(figsize=(10, 6))
df[num_cols].boxplot()
plt.xticks(rotation=45)
plt.show()
```

```python
Q1 = df[num_cols].quantile(0.25)
Q3 = df[num_cols].quantile(0.75)
IQR = Q3 - Q1

outliers = (df[num_cols] < (Q1 - 1.5 * IQR)) | (df[num_cols] > (Q3 + 1.5 * IQR))
outliers.sum()
```

```python
df['Outlier_Flag'] = outliers.any(axis=1).astype(int)
df['Outlier_Flag'].value_counts()
```

```python
for col in num_cols:
    lower = Q1[col] - 1.5 * IQR[col]
    upper = Q3[col] + 1.5 * IQR[col]
    df[col] = np.where(df[col] < lower, lower,
                       np.where(df[col] > upper, upper, df[col]))
```

```python
corr = df[num_cols].corr()
print(corr)
```

```python
df.to_csv("cleaned_dataset.csv", index=False)
print("Cleaned dataset saved successfully!")
```

```python
with open("eda_findings.txt", "w") as file:
    file.write("EDA Findings:\n")
    file.write("- Missing values handled using median/mode.\n")
    file.write("- Outliers detected using IQR method.\n")
    file.write("- Strong correlations observed between numeric variables.\n")
```
